```python
import numpy as np

docs = ['go india',
        'pakistan pakistan',
        'babar babar',
        'kohli kohli',
        'jeetega bhai jeetega pakistan jeetega',
        'winter is coming',
        'worst movie ever',
        'help ducky bhai',
        'babar azam scored a century',
        'virat and babar are goat']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences = pad_sequences(sequences, padding = 'post')
sequences
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Flatten, Embedding
```

```python
model = Sequential()
```

```python
#Embeding(# of unique words, output of embeding will give 2 numbers for each word, input length)
model.add(Embedding(23, output_dim = 2, input_length=5))
model.summary()
```

```python
model.compile('adam', ' accuracy')
```

```python
pred = model.predict(sequences)
print(pred)
```

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Flatten
```

```python
(X_train,Y_train),(X_test,Y_test) = imdb.load_data()
```

```python
X_train = pad_sequences(X_train, padding='post', maxlen=50)
X_test = pad_sequences(X_test, padding='post', maxlen=50)
```

```python
X_train.shape
```

```python
maxlen = 50        # sequence length
vocab_size = 25000 # example vocab size

model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=2, input_length=maxlen))
model.add(SimpleRNN(32))
model.add(Dense(1, activation='sigmoid'))

# Explicitly build the model:
model.build(input_shape=(None, maxlen))

model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
history = model.fit(X_train, Y_train, epochs = 5, validation_data = (X_test, Y_test))
```
